```python
import matplotlib.pyplot as plt
from random import uniform
from math import cos, sin, pi
import numpy as np
%matplotlib inline
```

```python
def generate1():  
    a = uniform(0, 1)  
    b = uniform(0, 1)  
    return (a * cos(2 * pi * b), a * sin(2 * pi * b))
def generate2():  
    while True:  
        x = uniform(-1, 1)  
        y = uniform(-1, 1)  
        if x ** 2 + y ** 2 > 1:  
            continue  
        return (x, y)
```

```python
first = np.array([generate1() for i in range(1000)])
```

```python
fig, ax = plt.subplots(1, 1)
ax.set_aspect('equal', 'box')
ax.plot(first[:,0], first[:, 1], '*', )
plt.show()
```

```python
second = np.array([generate2() for i in range(1000)])
fig, ax = plt.subplots(1, 1)
ax.set_aspect('equal', 'box')
ax.plot(second[:,0], second[:, 1], '*', )
plt.show()
```
